```python
def collapse_data_dictionary(
    df,
    view_col="View Name",
    field_col="Field",
    desc_col="Description",
    values_col="Values",
    sep=",",
    include_value_labels=True,
):
    """
    Collapse a nested data dictionary into one row per field.

    Assumes a 'header' row starts a block when either `view_col` or `field_col` is non-null.
    The header row's `desc_col` is the field description.
    Following rows list permissible values: `desc_col` = value label, `values_col` = code.
    """

    dd = df.copy()

    # Normalize blanks and literal 'NA'/'N/A' to real NaN
    for c in [view_col, field_col, desc_col, values_col]:
        dd[c] = (
            dd[c]
            .replace(r"^\s*$", pd.NA, regex=True)
            .replace({"NA": pd.NA, "N/A": pd.NA})
        )

    # Mark header starts (new field blocks) and assign group ids
    dd["_is_header"] = dd[view_col].notna() | dd[field_col].notna()
    dd["_gid"] = dd["_is_header"].cumsum()

    # Drop anything before the first header (gid == 0)
    dd = dd[dd["_gid"] > 0].copy()

    # Carry the most recent non-null View Name forward so blocks with only Field filled still get a view
    dd["_view_ffill"] = dd[view_col].ffill()

    def _agg_block(g):
        first = g.iloc[0]
        # Prefer the explicit view name on the header row; otherwise use forward-filled
        view = first[view_col] if pd.notna(first[view_col]) else first["_view_ffill"]
        field = first[field_col]
        field_desc = first[desc_col]

        # Permissible values (detail rows where code is present)
        val_rows = g[g[values_col].notna()]
        codes = [str(v) for v in val_rows[values_col].tolist()]
        labels = [str(v) for v in val_rows[desc_col].tolist()]

        out = {
            view_col: view,
            field_col: field,
            desc_col: field_desc,
            values_col: sep.join(codes) if codes else "",
        }
        if include_value_labels:
            out["Value Labels"] = sep.join(labels) if labels else ""

        return pd.Series(out)

    out = (
        dd.groupby("_gid", sort=True, as_index=False)
          .apply(_agg_block)
          .reset_index(drop=True)
    )

    # Keep only rows that truly represent fields
    out = out[out[field_col].notna()].reset_index(drop=True)
    return out

# --- Usage with a CSV ---
# Read your raw CSV into a DataFrame
#raw = pd.read_csv("raw_dictionary.csv")  # or specify dtype, encoding, etc.

# Collapse to one row per field
collapsed = collapse_data_dictionary(data_dict)
collapsed
# Save if you want
#collapsed.to_csv("collapsed_dictionary_collapsed.csv", index=False)
```
